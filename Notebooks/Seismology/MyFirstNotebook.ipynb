## Plotting functions

```python
# Initialize space in x and y
nx = 100
ny = 100

# initialize x
x = np.linspace(0,4*m.pi,nx)

# Initialize sin function
k = 10.

# Sin function
# f = np.sin(k*x)

# Gauss functions
a = 1. 
x0 = 2*np.pi
f = np.exp(-1./a**2  * (x -x0)**2) 

# Plot function
plt.plot(x,f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gauss function')
plt.show()
```
